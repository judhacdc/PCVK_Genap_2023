1. Implementasikan transformasi contrast pada Google Colaboratory menggunakan 
formula yang terdapat pada bagian Ulasan Teori

```python
print('Mengubah tingkat kecerahan citra')
print('--------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
  contrast = float(input('Masukkan Kontras [1.0-3.0]: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/Images/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

#access for each pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(contrast*original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

2. Implementasikan transformasi logarithmic brightness pada Google Colaboratory 
menggunakan formula yang terdapat pada bagian Ulasan Teori, 

```python
# non loop
print('Masukkan tingkat kecerahan dengan Transformasi Log ')
print('---------------------------------------------------')

try:
  brightness = int(input('Masukkan nilai kecerahan : '))
except ValueError:
  print('Error, not a number')


log_image = brightness * (np.log(1 + original))
log_image = np.array(log_image, dtype = np.uint8)

final_frame = cv.hconcat((original, log_image))
cv2_imshow(final_frame)
```

```python
#with loop
print(' Mengubah tingkat kecerahan citra dengan Transformasi Log ')
print('----------------------------------------------------------')

try:
  brightness = int(input('Masukan nilai kecerahan: '))
except ValueError:
  input('Error, not a number')

brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness * (np.log(1 + original[y,x,c])), 0 , 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

3. Implementasikan transformasi grayscale menggunakan metode averaging, lightness, 
dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada 
bagian Ulasan Teori

a. Averaging

```python
print('Averaging')
print('--------------------------------------------------------------------')

original = cv.imread('/content/drive/MyDrive/Images/mandrill.tiff')
original = cv.cvtColor(original,cv.COLOR_BGR2RGB)
gray_img = cv.imread('/content/drive/MyDrive/Images/mandrill.tiff')
gray_img = cv.cvtColor(gray_img,cv.COLOR_BGR2RGB)
R, G, B = original[:,:,0], original[:,:,1], original[:,:,2]
average = R/3+G/3+B/3
gray_img[:,:,0] = average
gray_img[:,:,1] = average
gray_img[:,:,2] = average
fig, axs = plt.subplots(1, 2)
axs[0].imshow(original)
axs[1].imshow(gray_img)
```

b. Lightness 

```python
print('Lightness')
print('--------------------------------------------------------------------')

original = cv.imread('/content/drive/MyDrive/Images/mandrill.tiff')
original = cv.cvtColor(original,cv.COLOR_BGR2RGB)
gray_img = cv.imread('/content/drive/MyDrive/Images/mandrill.tiff')
gray_img = cv.cvtColor(gray_img,cv.COLOR_BGR2RGB)
R, G, B = gray_img[:,:,0], gray_img[:,:,1], gray_img[:,:,2]
lightness = np.maximum(R,G,B)/2+ np.minimum(R,G,B)/2
gray_img[:, :, 0] = lightness
gray_img[:, :, 1] = lightness
gray_img[:, :, 2] = lightness
fig, axs = plt.subplots(1, 2)
axs[0].imshow(original)
axs[1].imshow(gray_img)
     
```

c.  Luminance

```python
print('Luminance')
print('--------------------------------------------------------------------')

original = cv.imread('/content/drive/MyDrive/Images/mandrill.tiff')
original = cv.cvtColor(original,cv.COLOR_BGR2RGB)
gray_img = cv.imread('/content/drive/MyDrive/Images/mandrill.tiff')
gray_img = cv.cvtColor(gray_img,cv.COLOR_BGR2RGB)
R, G, B = original[:,:,0], original[:,:,1], original[:,:,2]
luminance = 0.21*R+0.72*G+0.07*B
gray_img[:, :, 0] = luminance
gray_img[:, :, 1] = luminance
gray_img[:, :, 2] = luminance
fig, axs = plt.subplots(1, 2)
axs[0].imshow(original)
axs[1].imshow(gray_img)
```
