```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('/content/drive/MyDrive/Images/img01.jpg')

# Split channels
b, g, r = cv2.split(img)

# membuat channels kosong untuk blue dan green
empty_channel = 255 * np.ones_like(b)

# gabung chanel green dan blue
img_r = cv2.merge([g,empty_channel, empty_channel])

# menampilkan gambar
plt.imshow(img_r[5:30])
```
