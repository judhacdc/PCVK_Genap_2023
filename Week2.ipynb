```python
# Ubah Menjadi Loop untuk membuat persegi

# Membuat Black Image terlebih dahulu
black_img = np.zeros((512, 512, 3), dtype=np.uint8)

# Define the coordinates and colors of the rectangles
rectangles = [(50, 50, 200, 200, (0, 0, 255), 5),  # x1, y1, x2, y2, color, thickness
              (384, 0, 510, 150, (0, 255, 0), 10),
              (100, 300, 300, 450, (255, 0, 0), 10)]

# gambar persegi dengan menggunakan loop
for rect in rectangles:
    x1, y1, x2, y2, color, thickness = rect
    cv.rectangle(black_img, pt1=(x1, y1), pt2=(x2, y2), color=color, thickness=thickness)

diagonal = [(0,0, 512,512,(255,255,255),5)]
for diag in diagonal:
    x1, y1, x2, y2, color, thickness = diag
    cv.line(black_img, pt1=(x1, y1), pt2=(x2, y2), color=color, thickness=thickness)
    
# Tampilkan Gambar
plt.imshow(black_img)
plt.show()
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```
